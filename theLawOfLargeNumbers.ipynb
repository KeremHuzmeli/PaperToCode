```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
amp = 100

population_data = np.random.randn(N) * amp 

population_mean = np.mean(population_data)

plt.figure(figsize=(30,5))

plt.plot(population_data, "ko-", label="Population data")
plt.plot([0,N], [population_mean,population_mean], "r", linewidth=3, label="Population mean")
plt.xlabel("Count")
plt.ylabel("Value")
plt.title(f"The population mean is {round(population_mean,2)}")
plt.legend()
plt.show()
```

```python
epochs = 1000
sample_size = 100  

epochs_mean = np.zeros(epochs)

for i in range(epochs):
    sample = np.random.choice(population_data, size=sample_size)
    sample_mean = np.mean(sample)
    epochs_mean[i] = sample_mean

cumulative_mean = np.cumsum(epochs_mean) / (np.arange(epochs) + 1)

plt.figure(figsize=(15, 5))
plt.plot(cumulative_mean, "ko-", markersize=2, label="Cumulative Sample Mean")
plt.axhline(population_mean, color='r', linewidth=3, label="Population Mean")
plt.xlabel("Sample Iteration")
plt.ylabel("Mean Value")
plt.title("Law of Large Numbers - Convergence to Population Mean")
plt.legend()
plt.show()
```
