```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Xb=y
# (X^T X)^(-1) X^T X b = (X^T X)^(-1) X^T y
# b = (X^T X)^(-1) X^T y

X = np.array([[3000, 25],
              [3500, 30],
              [4000, 22],
              [4500, 35]])

y = np.array([500, 600, 700, 800])

X_T = X.T

X_T_X = np.dot(X_T, X)

X_T_X_inv = np.linalg.inv(X_T_X)

X_T_X_inv_X_T = np.dot(X_T_X_inv, X_T)

b = np.dot(X_T_X_inv_X_T, y)

b
```

```python
y_pred = np.dot(X, b)

y_pred
```

```python
plt.figure(figsize=(8, 6)) 
plt.plot(y, "ko", label="Real", markersize=6, alpha=0.7) 
plt.plot(y_pred, "ro", label="Predicted", markersize=6, alpha=0.7)  

plt.xlabel("Sample Index", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.title("Real vs Predicted Values", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
