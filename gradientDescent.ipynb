```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_derivation(x):
    return 2*x + 2

def get_equation(x):
    return x**2 + 2*x + 10
```

```python
x = np.linspace(-100, 100, 1000)

plt.plot(x, get_equation(x))
plt.grid(True)
plt.show()
```

```python
lr = 0.1
candidate_point = np.random.choice(x)

N = 100
points = np.zeros(N)
for i in range(N):

    deriv = get_derivation(candidate_point)
    candidate_point = candidate_point - lr*deriv
    points[i] = candidate_point


plt.plot(x, get_equation(x)) 
plt.plot(points, get_equation(points), "ro") 
plt.grid(True)
plt.show()
```
